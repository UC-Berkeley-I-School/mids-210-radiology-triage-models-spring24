```python
df = df.rename(columns={
    'tabular_classifications0': 'tabular_no_finding',
    'tabular_classifications1': 'tabular_atelectasis',
    'tabular_classifications2': 'tabular_cardiomegaly',
    'tabular_classifications3': 'tabular_lung_opacity',
    'tabular_classifications4': 'tabular_pleural_effusion',
    'notes_classifications0': 'notes_no_finding',
    'notes_classifications1': 'notes_atelectasis',
    'notes_classifications2': 'notes_cardiomegaly',
    'notes_classifications3': 'notes_lung_opacity',
    'notes_classifications4': 'notes_pleural_effusion',
    'img_classifications0': 'img_no_finding',
    'img_classifications1': 'img_atelectasis',
    'img_classifications2': 'img_cardiomegaly',
    'img_classifications3': 'img_lung_opacity',
    'img_classifications4': 'img_pleural_effusion',
})
```

```python
df['check_tabular'] = df['tabular_no_finding'] + df['tabular_atelectasis'] + df['tabular_cardiomegaly'] + df['tabular_lung_opacity'] + df['tabular_pleural_effusion'] > .99
df['check_notes'] = df['notes_no_finding'] + df['notes_atelectasis'] + df['notes_cardiomegaly'] + df['notes_lung_opacity'] + df['notes_pleural_effusion'] > .99
df['check_img'] = df['img_no_finding'] + df['img_atelectasis'] + df['img_cardiomegaly'] + df['img_lung_opacity'] + df['img_pleural_effusion'] > .99
```

```python
classification_names = ['no_finding', 'atelectasis', 'cardiomegaly', 'lung_opacity', 'pleural_effusion']
```

```python
df['max_tabular'] = df[['tabular_no_finding', 'tabular_atelectasis', 'tabular_cardiomegaly', 'tabular_lung_opacity', 'tabular_pleural_effusion']].idxmax(axis=1)
df['max_notes'] = df[['notes_no_finding', 'notes_atelectasis', 'notes_cardiomegaly', 'notes_lung_opacity', 'notes_pleural_effusion']].idxmax(axis=1)
df['max_img'] = df[['img_no_finding', 'img_atelectasis', 'img_cardiomegaly', 'img_lung_opacity', 'img_pleural_effusion']].idxmax(axis=1)
```

```python
df.columns
```

```python
df['max_tabular'] = df['max_tabular'].apply(lambda x: x.replace('tabular_', ''))
df['max_notes'] = df['max_notes'].apply(lambda x: x.replace('notes_', ''))
df['max_img'] = df['max_img'].apply(lambda x: x.replace('img_', ''))
```

```python
same_notes_img = df.loc[np.where((df['max_notes'] == df['max_img']) & (df['max_img'] != 'lung_opacity') & (df['max_img'] != 'no_finding') & (df['finding_names'] == df['max_img']))]
same_notes_img
```

```python
same_notes_img['max_notes_val'] = same_notes_img.apply(lambda x: x['notes_' + x['max_notes']], axis=1)
same_notes_img['max_img_val'] = same_notes_img.apply(lambda x: x['img_' + x['max_img']], axis=1)
```

```python
same_notes_img.reset_index(drop=True, inplace=True)
```

```python
same_notes_img.loc[np.where(same_notes_img['max_notes_val'] < same_notes_img['max_img_val'])][['patient_id', 'max_notes', 'max_notes_val', 'max_img', 'max_img_val', 'finding_names']]
```
